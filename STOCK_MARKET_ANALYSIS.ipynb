```python
import yfinance as yf 
```

```python
import pandas as pd 
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
stocks=['AAPL','MSFT','GOOGL','TSLA','AMZN']
df=yf.download(stocks,
               start='2023-01-01',
               end='2025-01-01'
              )
```

```python
df.info()
```

```python
df.describe()
```

```python
#data cleaning
```

```python
df=df.stack(level=1).reset_index()
```

```python
df.columns = [
    'Date',
    'Open',
    'High',
    'Low',
    'Close',
    'Adj_Close',
    'Volume'
]
```

```python
df.isnull().sum()  #missing values
```

```python
df.duplicated().sum()
```

```python
df[['Close','Volume']].describe() # check outliers
```

```python
plt.boxplot(df['Close'])
plt.show()
```

```python
# Exploratory Data Analysis
```

```python
#closing price trend
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Close'])
plt.title("Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
#Trending Volume Trend
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Volume'])
plt.title("Trading volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
df['Daily_Return'].hist(bins=50)
plt.title("Daily Return Distribution")
plt.show()
```

```python
#boxplot outliers
```

```python
plt.boxplot(df['Close'])
plt.title("close price outliers")
plt.show()
```

```python
df.to_csv("stock_market_cleaned.csv",
          index=False
         )
```
