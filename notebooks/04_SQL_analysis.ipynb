# 04 - SQL Database Analytics
This notebook demonstrates running SQL queries against database views to answer key analytical business questions.


```
import sys
import os
import pandas as pd
from sqlalchemy import text
sys.path.append(os.path.abspath('..'))
```

```
from config.database import get_engine
```

```
# Initialize database engine
engine = get_engine()
```

```
# Query 1: Rank companies by average close price
with engine.connect() as conn:
    df_avg = pd.read_sql(text('SELECT * FROM vw_average_close ORDER BY Avg_Close_Price DESC'), conn)
df_avg
```

```
# Query 2: Retrieve price extremes & trading volume averages
with engine.connect() as conn:
    df_extremes = pd.read_sql(text('SELECT * FROM vw_price_extremes'), conn)
df_extremes
```

```
# Query 3: Volatility and Risk Profile
with engine.connect() as conn:
    df_returns = pd.read_sql(text('SELECT * FROM vw_daily_returns_summary'), conn)
df_returns
```
